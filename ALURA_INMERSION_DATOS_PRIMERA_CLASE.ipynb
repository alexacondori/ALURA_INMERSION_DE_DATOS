2. CONSULTAR OTROS DATOS ESTADISTICOS,CONTEO,MEDIANA,VALORES MINIMO Y MAXIMO

```python
#agrupa por columnas
promedio_areas_barrio = inmuebles.groupby('Barrio')['Area'].mean()
promedio_areas_barrio
```

```python
promedio_areas_barrio.sample(10)
```

```python
import matplotlib.pyplot as plt
ax=promedio_areas_barrio.head(10).plot.bar(color='pink')
ax.set_title('Promedio de Áreas por Barrio')  # Agrega un título al gráfico
plt.xlabel('Barrios')
plt.ylabel('Promedio de Áreas')
plt.show()# Muestra el gráfico
```
